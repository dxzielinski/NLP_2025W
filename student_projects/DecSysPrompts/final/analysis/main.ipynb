## Regulative statements analysis

The distribution of these indicators across selected providers is shown in Figure 10. Anthropic and OpenAI show a strong preference for Strategy statements, indicating that these providers rely on a *descriptive* rather than a *direct* governance style, which may increase the likelihood of hallucinations.


```python
plot_regulative_statements_heatmap(prompts_analysis)
```

Meta adopts a strongly restrictive governance style dominated by prohibitions. In contrast, OpenAI, Google, and Anthropic rely primarily on commands with moderate prohibitions and minimal permissions. Overall, xAI exhibits the most diverse governance structure, balancing prohibitions, commands, and permissions, while the remaining providers follow more directive or restrictive approaches.
